```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# enable IPython to display matplotlib graphs
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
```

```python
my_plot = df.plot(x='N', y='Mflops', kind='line', logx=True, grid=True, )
my_plot.set_xlabel('lgN')
my_plot.set_ylabel('MFlops')
```

```python
df['GBs'] = df['Mflops'] * 16 / 1000
```

```python
df['GBs']
```

```python
my_plot = df.plot(x='N', y='GBs', kind='line', logx=True, grid=True, )
my_plot.set_xlabel('lgN')
my_plot.set_ylabel('Bandwidth (GB/s)')
```

```python
df.drop(['MBs'], axis=1)
```
